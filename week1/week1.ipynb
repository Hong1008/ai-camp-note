```python
# 게임의 규칙은 다음과 같습니다
# 숫자 1부터 시작하며, 마지막으로 31을 말하는 사람이 패배합니다.
# 각 플레이어는 한 번에 1개에서 3개의 숫자를 말할 수 있습니다.
# 플레이어가 말한 숫자는 이전 숫자에 이어야 하며, 연속된 숫자를 말해야 합니다.
# 예: 이전 숫자가 5라면, 6, 6과 7, 또는 6, 7, 8을 말할 수 있습니다.
# 31을 말한 사람이 패배합니다.

# 게임 시작!
# 숫자 몇 개를 부를 건가요? (1~3):2
# 벌칙 당첨!!

print('게임 시작!')

def call_input():
     i = input('숫자 몇 개를 부를 건가요? (1~3):')
     if (not(i.isnumeric)):
          print('숫자를 입력하세요.')
          return call_input()
     i = int(i)
     if (i > 3):
          print('3이하의 숫자만 입력해주세요.')
          return call_input()
     return i 

num = 0
while num < 31:
     i = call_input()
     for n in range(num, num + i):
          num += 1
          if (num > 31):
               print('벌칙 당첨!!')
               break
          print(f'{num}!')
     
          
```

```python
# 문제
# 1에서부터 6까지의 눈을 가진 3개의 주사위를 던져서 다음과 같은 규칙에 따라 상금을 받는 게임이 있다.

# 같은 눈이 3개가 나오면 10,000원+(같은 눈)×1,000원의 상금을 받게 된다. 
# 같은 눈이 2개만 나오는 경우에는 1,000원+(같은 눈)×100원의 상금을 받게 된다. 
# 모두 다른 눈이 나오는 경우에는 (그 중 가장 큰 눈)×100원의 상금을 받게 된다.  
# 예를 들어, 3개의 눈 3, 3, 6이 주어지면 상금은 1,000+3×100으로 계산되어 1,300원을 받게 된다.
#  또 3개의 눈이 2, 2, 2로 주어지면 10,000+2×1,000 으로 계산되어 12,000원을 받게 된다. 
#  3개의 눈이 6, 2, 5로 주어지면 그 중 가장 큰 값이 6이므로 6×100으로 계산되어 600원을 상금으로 받게 된다.

# N(2 ≤ N ≤ 1,000)명이 주사위 게임에 참여하였을 때, 가장 많은 상금을 받은 사람의 상금을 출력하는 프로그램을 작성하시오.

# 입력
# 첫째 줄에는 참여하는 사람 수 N이 주어지고 그 다음 줄부터 N개의 줄에 사람들이 주사위를 던진 3개의 눈이 빈칸을 사이에 두고 각각 주어진다. 

def prize(pips: list[int]) -> int:
    a = set(pips)

    if len(a) == 1:
        return 10000 + 1000*pips[0]
    elif len(a) == 2:
        pip = sum(pips) - sum(a)
        return 1000 + 100 * pip

    return 100 * max(pips)

N = int(input())

pips = [[int(i) for i in input().split()] for _ in range(N)]

print(max([prize(p) for p in pips]))
```

```python
def solution(polynomial:str):
    str_list = polynomial.split(' + ')

    my_int = sum([int(s) for s in str_list if s.isdigit()])

    x_cnt = sum([1 if s[:-1] == '' else int(s[:-1]) for s in str_list if s.endswith('x')])


    return (f'{"" if x_cnt == 1 else x_cnt}x' if x_cnt >= 1 else '') + ('' if my_int == 0 or x_cnt == 0 else ' + ') + (f'{my_int}' if my_int > 0 else '')

print(solution('1'))
print(solution('x + 3x'))
print(solution('1 + 3'))
print(solution('x + 3'))
```

```python
import re
from konlpy.tag import Okt
from collections import Counter

# 1. 파일 읽기
with open("./week1/content/bts_lyrics_top50.txt", "r", encoding="utf-8") as f:
    text = f.read()

# 2. 정규표현식으로 한글만 남기기 (영어 가사를 포함하고 싶다면 이 단계 생략)
# text = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s]', ' ', text)

# 3. 형태소 분석 및 명사 추출
okt = Okt()
nouns = okt.nouns(text)

# 4. 불용어(Stopwords) 제거
# 가사 특성상 자주 나오지만 의미 없는 단어들을 추가하세요.
stopwords = ['내', '나', '난', '날', '너', '널', '것', '수', '등', '거', '의', '때', '니']
words = [n for n in nouns if n not in stopwords and len(n) > 1]

# 5. 빈도수 계산
word_counts = Counter(words)

print(word_counts)
```

```python
masking_image = np.array(Image.open('./week1/content/bts-logo.jpg'))

wordcloud = WordCloud(font_path='./week1/content/NanumBarunGothic.ttf',
                mask=masking_image,
                width=1200,
    height=800, max_font_size=100, 
                 colormap='PuRd',      # BTS 느낌이 나는 보라색 계열 테마 추천
                max_words=100,
                 background_color='#0E0E0E').generate_from_frequencies(word_counts)

pit.imshow(wordcloud)
pit.axis('off')
pit.figure(figsize=(15, 10))
pit.imshow(wordcloud, interpolation='bilinear')
pit.tight_layout(pad=0)
pit.show()
```
